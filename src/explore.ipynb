### Clean the generated table 

```python

# TODO: List the columns we want to keep in the following list, `columns_to_keep`
columns_to_keep = ["movie_id","title","overview","genres","keywords","cast","crew"]
total_data = total_data[columns_to_keep]
total_data.head()
```

### Transform the data

```python

# Check the data type of `genres`, `keywords`, `cast`, `crew`
columns_to_check = ["genres", "keywords", "cast", "crew"]

for column in columns_to_check:
    print(f"Data type of the first row for column `{column}`: {type(total_data.loc[0, column])}")
```

```python
import ast

# TODO: Convert string representation of lists to actual lists
total_data['genres'] = total_data['genres'].apply(ast.literal_eval)
total_data['keywords'] = total_data['keywords'].apply(ast.literal_eval)
total_data['cast'] = total_data['cast'].apply(ast.literal_eval)
total_data['crew'] = total_data['crew'].apply(ast.literal_eval)
# Verify that the data type was casted from a str to list
for column in columns_to_check:
    print(f"Data type of the first row for column `{column}`: {type(total_data.loc[0, column])}")
```

```python

# TODO: Complete the `extract_name()` function to extract the 'name' attribute from the JSON
def extract_name(json_list):
    names = [item['name'] for item in json_list]
    return names

# Apply the `extract_name` function to 'genres' and 'keywords' columns
total_data['genres'] = total_data['genres'].apply(extract_name)
total_data['keywords'] = total_data['keywords'].apply(extract_name)

# TODO: Complete the `select_first_three_names()` function to select the first three names from the 'cast' column

def select_first_three_names(json_list):
    names = [item['name'] for item in json_list[:3]]
    return names
    

# Apply the `select_first_three_names` function to 'cast' column
total_data['cast'] = total_data['cast'].apply(select_first_three_names)

total_data.head()
```

```python

first_row_crew = total_data.loc[0, 'crew']

for dict_obj in first_row_crew:
    print(dict_obj)
    print()
```

```python
# Define the select_director_name function to select the director's name from the crew list
def select_director_name(json_list):
    name = ''
    # Go through each dictionary in list to find the director's name
    for dict_obj in json_list:
        if dict_obj["job"] == "Director":
            name = dict_obj["name"]
    return name

# Apply the select_director_name function to 'crew' column
total_data['crew'] = total_data['crew'].apply(select_director_name)

total_data.head()
```

```python
# Convert the 'genres', 'keywords', and 'cast' columns to strings without adding separators between elements
# Convert the 'genres', 'keywords', and 'cast' columns to strings without adding separators between elements
total_data['genres'] = total_data['genres'].apply(lambda x: ''.join(map(str, x)))
total_data['keywords'] = total_data['keywords'].apply(lambda x: ''.join(map(str, x)))
total_data['cast'] = total_data['cast'].apply(lambda x: ''.join(map(str, x)))

total_data.head()

```

```python
total_data[total_data.isnull().any(axis=1)]
```

```python
# fill missing value with string
total_data.fillna('', inplace=True)
```

```python
# Combine the columns into a single column called 'tags'
total_data['tags'] = total_data.apply(lambda row: ''.join(row['overview']) + '' + ''.join(row['genres']) + '' + ''.join(row['keywords']) + '' + ''.join(row['cast']), axis=1)

# Print the 'tags' column for the first row
print(total_data.iloc[0].tags)

# Display the updated dataframe
total_data.head()
```

Vectorizing

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english')
vectorized_tags = vectorizer.fit_transform(total_data['tags'].values.astype('U'))

print (vectorized_tags)
```

### Fitting Model

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectorized_tags)

# Define the recommend function
def recommend(movie):
    movie_index = total_data[total_data["title"] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    recommended_movies = [total_data.iloc[i[0]]['title'] for i in movie_list]
    return recommended_movies

# Example usage
recommended_movies = recommend("Avatar")
print("Recommended movies for Avatar:", recommended_movies)
```
